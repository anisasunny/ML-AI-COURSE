```python
import pandas as pdas
import numpy as nump
import matplotlib.pyplot as matplt
import seaborn as sborn
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
data = pdas.read_csv('/content/WineQT.csv')
print(data.head())

# Check dataset dimensions
print(f"Dataset Shape: {data.shape}")

# Check column data types & missing values
print("\nData Info:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())

print("\nQuality Distribution:")  # wine quality counts
print(data["quality"].value_counts())
# Visualizations

#plot histogram

for cl in data:
    matplt.figure(figsize=(8, 6))
    sborn.histplot(data[cl], kde=True)
    matplt.title(f'Summary of {cl}')
    matplt.show()
#box plots
for col in data:
    matplt.figure(figsize=(8, 6))
    sborn.boxplot(y=col, data=data)
    matplt.title(f'Box Plot of {col}')
    matplt.show()
#density plot
for col in data:
    matplt.figure(figsize=(8, 6))
    sborn.kdeplot(data[col])
    matplt.title(f'Density Plot of {col}')
    matplt.show()
#scatter plot

matplt.figure(figsize=(10, 6))
sborn.scatterplot(x='alcohol', y='volatile acidity', hue='quality', data=data)
matplt.title('Scatter Plot: Alcohol vs.volatile acidity')
matplt.show()
#pair plot
sborn.pairplot(data, hue='quality')
matplt.show()
#bar charts
sborn.countplot(x='quality', data=data)
matplt.title('Distribution of Wine Quality Labels')
matplt.show()
#count plot
sborn.countplot(x='quality', data=data)
matplt.title('Summary of Wine Quality')
matplt.show()
```
